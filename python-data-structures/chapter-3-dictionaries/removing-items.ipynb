##### Removing Items with `del`

```python
sensors = {
    "temperature": "22°C",
    "humidity": "60%",
    "pressure": "1013 hPa"
}

# If the key doesn't exist, del will raise a KeyError
try:
    del sensors["humidity"]
except KeyError:
    print("Humidity doesn't exist in dict sensors")

print(sensors)
```

##### Removing Items with `pop()`

```python
sensors = {
    "temperature": "22°C",
    "humidity": "60%",
    "pressure": "1013 hPa"
}

# pop() will return the value of the removed element
value = sensors.pop("pressure")
# You can set the send value as the default value

print(value)
print(sensors)
```

##### Removing Items with `popitem()`

```python
sensors = {
    "temperature": "22°C",
    "humidity": "60%",
    "pressure": "1013 hPa"
}

# popitem() will remove and return the last key-value pair
last_element = sensors.popitem()

print(last_element)
print(sensors)
```

##### Clearing the Dictionary with `clear()`

```python
sensors = {
    "temperature": "22°C",
    "humidity": "60%",
    "pressure": "1013 hPa"
}

sensors.clear()
print(sensors)
```
