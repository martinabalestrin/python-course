##### Creating Dictionary View Objects with `items()`

```python
prices = {
    'keyboard': 29.99,
    'monitor': 189.99,
    'mouse': 19.99,
    'chair': 120.00
}

print(prices.items())
```

##### Sorting Dictionaries by Key

```python
prices = {
    'keyboard': 29.99,
    'monitor': 189.99,
    'mouse': 19.99,
    'chair': 120.00
}

alphabetical = dict(sorted(prices.items()))
print(alphabetical)
```

##### Sorting Dictionaries by Value

```python
prices = {
    'keyboard': 29.99,
    'monitor': 189.99,
    'mouse': 19.99,
    'chair': 120.00
}

# item[1] from the key argument refers to the view object returned from items()
sorted_prices = dict(sorted(prices.items(), key=lambda item: item[1]))
print(sorted_prices)
```

##### Sorting Dictionaries by Value with `itemgetter()`

```python
from operator import itemgetter

prices = {
    'keyboard': 29.99,
    'monitor': 189.99,
    'mouse': 19.99,
    'chair': 120.00
}

# Item
sorted_prices_getter = dict(sorted(prices.items(), key=itemgetter(1)))
print(sorted_prices_getter)
```

`operator.itemgetter()` operates faster than `lambda` because it's implemented in C.
