##### Creating Dictionary View Objects with `values()`

```python
uptime_hours = {
    'server1': 120,
    'server2': 98,
    'server3': 143,
    'server4': 0 # offline
}

print(uptime_hours.values())
values = uptime_hours.values()

print('Total Servers:', len(values))
print('Total uptime:', sum(values))
print('Max uptime:', max(values))
print('Min uptime:', min(values))
print('All running:', all(values))
print('Any running:', any(values))
```
