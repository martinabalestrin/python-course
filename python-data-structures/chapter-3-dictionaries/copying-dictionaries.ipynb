##### Aliasing Dictionaries

```python
original = {"theme": "dark"}

alias = original

alias["theme"] = "light"

# Both Alias and Original point to the same space in memory
print(original)
print(alias)
```

##### Making a Shallow Copy of a Dictionary with `copy()`

```python
original = {"theme": "dark"}

shallow_copy = original.copy()
# You can also use dict()
# shallow_copy = dict(original)

shallow_copy["theme"] = "light"

print(original)
print(shallow_copy)
```

##### Making a Deep Copy of a Dictionary with `deepcopy()`

```python
import copy

original = {
    "theme": "dark",
    "options": {
        "autosave": True
    }
}

deep_copy = copy.deepcopy(original)

deep_copy["options"]["autosave"] = False

print(original)
print(deep_copy)
```
