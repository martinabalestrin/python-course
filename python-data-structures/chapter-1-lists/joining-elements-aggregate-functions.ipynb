##### Joining List Elements into a String

```python
words = ["Hi!", "I'm", "Martina"]

phrase = " ".join(words)

print(phrase)
```

##### Aggregate Functions

```python
# Sensor readings
readings = [0.0, 0.1, 0.0, 0.2]

print('Sum:', sum(readings))
print('Maximum:', max(readings))
print('Minimum:', min(readings))
print('Any:', any(readings)) # if at least any element is truthy
print('All:', all(readings)) # if all elements are truthy
```
