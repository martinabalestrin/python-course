##### Counter Dictionaries

```python
from collections import Counter

letters = ['a', 'a', 'b', 'c', 'a', 'c']

c = Counter(letters)
print(c)
```

##### Accessing Counts

```python
from collections import Counter

# Simulated voting poll
votes = [
    "Python", "Java", "Python", "Rust", "Java", "Python"
]

vote_counts = Counter(votes)

print("Votes for Python:", vote_counts["Python"])
```

##### Updating Counts

```python
from collections import Counter

# Simulated voting poll
votes = [
    "Python", "Java", "Python", "Rust", "Java", "Python"
]

vote_counts = Counter(votes)

new_votes = [
    "Java", "Java", "Rust", "Java"
]

vote_counts.update(new_votes)
# You can use .substract() to do the opposite

print(vote_counts)
```

##### `Counter` methods:

`most_common()`: returns the n highest-count items

`total()`: returns all counts

`elements()`: expands the counts into a list (reverse of Counter)

##### Demo Anagram Implementation with `Counter`

```python
from collections import Counter

def is_anagram(a, b):
    return Counter(a) == Counter(b)

print(is_anagram("amor", "roma"))
```
