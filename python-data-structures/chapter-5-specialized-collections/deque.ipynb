##### Creating a `deque` 

```python
from collections import deque

# Creating an empty deque
dq = deque()
print(dq)

dq = deque((1,2,3)) # tuple
print(dq)

dq = deque([1,2,3]) # list
print(dq)

dict = {"a": 1, "b": 2, "c": 3}
dq = deque(dict.items()) # dictionary
print(dq)
```

##### Appending and Popping Elements form Both Sides

```python
from collections import deque

dq = deque([1,2,3]) 

# Pop and Append normal on the right
dq.pop()
dq.append(4)
print(dq)

# Pop and Append + 'left' for left
dq.popleft()
dq.appendleft(5)
print(dq)
```

##### Commands with `deque`

`maxlen`: can be used as a second parameter on `deque` to indicate the maximum of elements on a deque. When it's in maximum capacity, if you append something else, a element on the otherside will be discarted.

`rotate()`: rotates the elements in the deque by a specified number of steps to the right (default 1).
