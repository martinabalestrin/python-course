##### Creating Sets

```python
# Set elements have to be immutable
# Sets are unordered
my_set = {1, 2, 3, "s1", "s2"}
print(my_set)

# Create empty sets with the set() constructor
empty_set = set()
print(empty_set)

# Empty curly braces would create a dictionary
# empty_dict = {}
```

##### Set Elements are Unique

```python
# Set will remove all duplicate elements
my_set = {1, 1, 2, 2, 3, 3}
print(my_set)

# You can initialize sets with elements from other iterable structures
my_tuple = ("a", "b", "c")
set_from_tuple = set(my_tuple)
print(set_from_tuple)

# This can be useful for removing duplicate elements
visitor_id_list = ["user123", "user456", "user123", "user101", "user456"]
unique_visitors_set = set(visitor_id_list)
print(unique_visitors_set)
```

##### Sets and Strings

```python
# Strings are iterable as well
strings_set = set("Hello")
print(strings_set)

# If you want to store a full string as one item
strings_set = {"Hello"}
print(strings_set)
```

##### Aggregate Functions

```python
numbers_set = {3, 7, 2, 9, 5}

print(len(numbers_set))
print(sum(numbers_set))
print(min(numbers_set))
print(max(numbers_set))
print(any(numbers_set))
print(all(numbers_set))
```
