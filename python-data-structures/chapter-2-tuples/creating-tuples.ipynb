##### Creating Tuples

```python
empty_tuple = ()

# Comma after the number is required, otherwise Python will think it's an integer
one_element_tuple = (22,)
print(one_element_tuple)

multiple_elements = (1, 'banana', 3.14)
print(multiple_elements)
print('Length:', len(multiple_elements))

# Tuple without parentheses - it's just necessary to have commas between the values
implicit_tuple = 10, 20, 30
print(implicit_tuple)
```

##### Accessing Tuple Elements

```python
# RGB color for a soft purple
rgb_color = (150, 100, 200)

print(rgb_color)
print('Red Component:', rgb_color[0])
print('Green Component:', rgb_color[1])
print('Blue Component:', rgb_color[2])
```
